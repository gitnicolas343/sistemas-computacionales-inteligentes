```python
# TODO
variables = ['WA','NT','SA','Q','NSW','V','T']
colores = ['R','G','B']
dominios = {'WA':[colores[1]],'NT':colores.copy(),'SA':colores.copy(),'Q':colores.copy(),'V':[colores[0]],'T':colores.copy(), 'NSW':colores.copy()}  #con la asignacion parcial WA = GRNN , V =  RED
vecinos = {'WA': ['NT', 'SA'],'NT': ['WA', 'SA', 'Q'],'SA': ['WA', 'NT', 'Q', 'NSW', 'V'],'Q': ['NT', 'SA', 'NSW'],'NSW': ['SA', 'Q', 'V'], 'V': ['SA', 'NSW'],'T': []
}
```

```python
restricciones = list()   #AQUI HAGO las conbinaciones que resultan en las restricciones del problema

for i in vecinos:
    for j in vecinos[i]: #itero sobre el diccionario sobre claves y valores de las listas (diccionario de listas)
        restricciones.append((i,j))  #agrego como tuplas
```

```python
#revisar(A,B)
def revision(domains, X_i, X_j):        # funcion revise necesaria en AC3
                                       #iniciamos en false segun el pseudo codigo
   revised = False
   for x in domains[X_i].copy():       # iteramos sobre el dominio de la X_i(uso una copia para no modificar la info original, salian errores y generalmente la primera ejecucion daba una respuesta y la segunda otra )
       if x in domains[X_j]:
           if len(domains[X_j]) == 1:  #garantia de que ya hemos acabado con el dominio y solo queda une elemento en x_j
               domains[X_i].remove(x) # se elemina x de x_i
               revised = True
   return revised
```

```python
def AC3(dominios,arco,vecinos): #segui el psudocodigo de clase que la verdad salio casi directo en python
    cola = arco.copy()           #hago una copia( me di cuenta por las malas que si no se hace copia la lista  salen incosistencias)
    while cola:
        (X_i, X_j) = cola.pop(0)                              #esta estructura me permite ir quitando elemento inicial y ademas sacar los elemento de la tupla
        if revision(dominios,X_i,X_j):                        # se evalua la funcion revise
            if not dominios[X_i] :                            # se verifica si hay algo en el dominio de x_i
                return False
                                                             # si hay algo
            for z in vecinos[X_i]:                           # iteramos sobre los vecinos
                if z != X_j:
                    cola.append((z,X_i))
    return True
```

```python
result = AC3(dominios,restricciones,vecinos)

print('es cosistente?', result)
print('dominios',dominios)
```
